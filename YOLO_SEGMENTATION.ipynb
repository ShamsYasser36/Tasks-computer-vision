```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load a model (YOLOv8)
model = YOLO("yolov8n-seg.pt")  


results = model("photo_5936203136391824702_x.jpg")  

result_img = results[0].plot()
plt.imshow(result_img)
plt.axis('off')
plt.show()
```
